# This Notebook is to show off the functionality of the APR class

The APR class is for loading in APR2/APR3 files into an xarray dataset. The example here is using a file from olympex, but the file has been thinned inorder to meet github's filesize requirement. 

Note: GCPEX files require hdf4 readers, which I have not tested here. I think the package I have is this one: https://github.com/fhs/pyhdf

```python
import drpy
%pylab inline
```

```python
filename = '../example_file/olympex_apr3_20151203_150904_R2_KUsKAsWs.h5'
apr = drpy.core.APR()
apr.read(filename=filename,campaign='olympex')
```

```python
plt.pcolormesh(apr.xrds.time3d[:,12,:],apr.xrds.alt3d[:,12,:],apr.xrds.Ku[:,12,:],cmap=drpy.graph.cmaps.HomeyerRainbow,vmin=0,vmax=40)
plt.colorbar()
```
